```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

oe is an object of ordinal encoder nd then pass 1 parameter which is going to be our list i hv passed 2 lists 1 is result and 1 is education,in that list i hv passed hw many categorical data has to be transformed

if i dnt give categories there it will randmly decide values to be specified

```python
oe.fit(X_train) #fit that in train
```

```python
X_train = oe.transform(X_train) #transforming both x test and y test
X_test  = oe.transform(X_test)
```
