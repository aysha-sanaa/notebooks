```python
pump = 0

while True:
    up = int(input("Enter up sensor value : "))
    down = int(input("Enter down sensor value : "))
    
    if up == 0 and down ==0:
        pump = 1
    
    elif up ==1 and down ==1:
        pump = 0
        
    elif up == 0  and down ==1:
        pump = pump
        
    else:
        print("Invalid")
        break
    
    print(f"{pump = }")
    

```
