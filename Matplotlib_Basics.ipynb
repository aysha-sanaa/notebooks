```python
x=np.arange(5)
y=x**3
plt.plot([1,2,3,4],[1,4,9,16],"go",x,y,"r^")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")

plt.subplot(1,2,2)
plt.plot(x,y,"r^")

plt.suptitle("Sub-plots")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")

plt.subplot(2,1,2)
plt.plot(x,y,"r^")

plt.suptitle("Sub-plots")
plt.show()
```

```python
x=np.arange(1,5)
y=x**3
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(6,6))
ax[0,0].plot([1,2,3,4],[4,3,2,1],"k")
ax[0,1].plot([1,2,3,4],[1,4,9,16],"r")
ax[1,0].plot(x,y,"g^")
ax[1,1].plot([1,2,3,4],[3,6,9,12],'b')
ax[0,1].set_title("Squares")
ax[1,0].set_title("Cubes")
plt.show()
```

```python
div=["a","b","c","d"]
div_marks=[70,80,65,90]
plt.bar(div,div_marks,color="green")
plt.show()
```

```python
div=["a","b","c","d"]
div_marks=[70,80,65,90]
boy_marks=[60,80,75,95]

index=np.arange(4)
width=0.4

plt.bar(index,div_marks,width,color='red',label='div-marks')
plt.bar(index+width,boy_marks,color='blue',label='boy-marks')

plt.xticks(index+ width/2,div)

plt.legend(loc='best')
plt.show()
```

```python
firms=["a","b","c","d","e"]
market=[20,25,15,10,20]
explode=[0,0.5,0,0,0]

plt.pie(market,explode=explode,labels=firms,shadow=True,startangle=45)
plt.axis('equal')

plt.legend(title="firms")
plt.show()
```
